```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['font.family'] = 'NanumBarunGothic'  # (Colab 용) 한글 출력을 위한 글꼴 설정
import numpy as np
import pandas as pd
```

```python
# 일자별 주식 데이터 불러오기 (dict, for)
d = {}

# for문을 이용해서 데이터 읽기
  # adj_close : 주가
  # mc : 시가총액
  # sales : 매출액
  # op : 영업이익
  # ni : 순이익
  # liab : 부채
  # eq : 자본
for data_name in ['adj_close', 'mc', 'sales', 'op', 'ni', 'liab', 'eq'] :
  print(f"read csv {data_name} ... ")
  d[data_name] = pd.read_csv(f"./stock.{data_name}.csv", index_col=0, encoding="cp949", parse_dates=True)
  print(f"done! {d[data_name].shape}")

## result : 2278 개의 주식 데이터, 총 582일간 기록함
```

```python
# 전체 주식시장에서 산업별로 차지하는 비중 계산
# 분석을 수행할 날짜 지정
view_date = "2020-10-14"
```

```python
# concat할 데이터 살펴보기 
# 비중을 확인해야하므로 시총 확인
d['mc'].loc[:, view_date]
```

```python
# 업종과 해당 날짜의 시총, 두개의 데이터 프레임을 병합
name_sector_df = pd.concat([stock_info, d['mc'].loc[:, view_date]], axis=1)
name_sector_df
```

```python
# query를 사용해서 가장 많이 차지하는 섹터에 어떤 종목이 있는지 확인
name_sector_df.query('Sector == "하드웨어"')
```

```python
# 최근 수익률이 좋았던 종목 확인
  # pct_change 함수로 변동률 구하기
  # 100 을 곱해줘야 백분율 데이터가 나옴
# 데이터 확인
d['adj_close'].head()
```

```python
# 1일간의 주가 변동률 확인
d['adj_close'].pct_change(periods=1, axis=1) * 100
```

```python
# 3일간의 주가 변동률 확인
d['adj_close'].pct_change(periods=3, axis=1) * 100
```

```python
# 여러투자 기간의 수익률을 계산하여 딕셔너리 d에 할당
  # 주식 시장이 주말에는 쉬는걸 감안해서 계산
for hold_days in [1, 3, 5, 10, 20, 60, 120] :
  table_name = f'return_{hold_days}d'
  d[table_name] = d['adj_close'].pct_change(periods=hold_days, axis = 1) * 100

  print(f'투자기간 {hold_days}일, 테이블 {table_name} 연산 완료 !')
```

```python
# 딕셔너리 d 확인
d['return_20d']
```

```python
# 데이터프레임을 입력하면, stock_info를 앞에 붙여주는 함수 생성
def get_df_with_info(data_df) :
  return pd.concat([stock_info, data_df], axis=1)
```

```python
# 가설과 실험 시작
# 전체 날짜 리스트 만들기
date_list = d['adj_close'].columns
date_list
```

```python
# 적자 종목 제거, 하락 추세 종목 제거, 종목별 per 산출, 종목별 거래대금 분석 등 데이터를 가공해서 종목 선별
# 20일간의 투자 수익률을 계산해야하므로 뒤에서 20번째 날짜를 골라야함
date_list[-21], date_list[-1]
```

```python
# 매수할 날짜 설정
buy_date = date_list[-21]
```

```python
# per, pbr 계산해서 딕셔너리에 저장
d['per'] = d['mc'] / d['ni']    # 시가총액 / 순이익
d['pbr'] = d['mc'] / d['eq']    # 시가총액 / 자본
```

```python
# 매수 날짜에 해당하는 데이터 추출
series_list = []
for table_name in ['adj_close', 'mc', 'sales', 'op', 'ni', 'return_20d', 'per', 'pbr'] :
  sr = pd.Series(d[table_name].loc[:, buy_date])
  sr.name = table_name
  series_list.append(sr)
```

```python
# 추출한 serires들로 매수 날짜의 데이터프레임 만들기
date_data = pd.concat(series_list, axis=1)
date_data
```

```python
# 필터링을 우량한 조건을 걸어 종목 선별
  # ni > 0 : 순이익 흑자
  # return_20d > 1.0 : 이전 20일 기준 수익률 1% 이상
  # per < 10 : per 10 이하
  # pbr < 10 : pbr 10 이하
date_data.query("ni > 0 and return_20d > 1.0 and per < 10 and pbr < 10")
```

```python
# 선별된 종목의 인덱스 추출
buy_stock_index = date_data.query("ni > 0 and return_20d > 1.0 and per < 10 and pbr < 10").index
buy_stock_index
```

```python
# 분석을 통해 선별한 종목에 투자한 경우 수익률 확인
# 전 종목의 평균 수익률
d['return_20d'].loc[:, '2020-10-14'].mean()
```

```python
# 선별한 종목의 평균 수익률
d['return_20d'].loc[buy_stock_index, '2020-10-14'].mean()
```

```python
# 섹터 기반해서 종목 선정 후 수익률 분석
sector_20d_df = get_df_with_info(d['return_20d'].loc[:, buy_date]).groupby('Sector').mean()
sector_20d_df
```

```python
sector_20d_df.sort_values(buy_date, ascending=False)
```

```python
# 매수 날짜 데이터 프레임에 섹터를 추가한 데이터 프레임 생성
date_data_with_sector = get_df_with_info(date_data)
date_data_with_sector
```

```python
# 필터링 조건
  # 이전 20일 기준 수익률이 가장 높은 섹터
  # pbr < 10
  # ni > 0
date_data_with_sector.query("Sector == '에너지' and pbr < 10 and ni > 0")
```

```python
# 선별된 종목 인덱스 추출
buy_stock_index = date_data_with_sector.query("Sector == '에너지' and pbr < 10 and ni > 0").index
buy_stock_index
```
