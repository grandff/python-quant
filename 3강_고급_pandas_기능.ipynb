```python
# 일자별 주가정보 파일
adj_close = pd.read_csv("stock.adj_close.csv", index_col=0, encoding="cp949")
```

```python
# 주식 코드별 기업명, 산업분류코드
## 한글이 포함되어있는 경우 인코딩 옵션은 필수고, cp949를 사용함
itemname = pd.read_csv("stockinfo.itemname.csv", index_col=0, encoding="cp949")
sector = pd.read_csv("stockinfo.sector.csv", index_col=0, encoding="cp949")
```

```python
# 기업명과 산업분류코드 데이터프레임 하나로 합치기
stock_info = pd.concat([itemname, sector], axis=1)
```

```python
# 14개의 주식정보만 가져오기 (상단에서)
df_a = stock_info.iloc[:14]
df_a
```

```python
# 2020년 10월 13일, 14일 두 날짜의 14개 주식정보 가져오기 
df_b = adj_close.loc[:, '2020-10-13':].iloc[:14]
df_b
```

```python
# concat을 사용해 위에서 만든 두 dataframe 합치기
# axis = 1 옵션 필수. 안줄경우 이상하게 합쳐짐.
my_concat_df = pd.concat([df_a, df_b], axis=1)
my_concat_df
```

- Groupby를 사용해서 Aggregation

```python
# sector 영역별로 sum 계산
my_concat_df.groupby('Sector').sum()

# 평균값 mean, 갯수 count, 최댓값 max
```

- query로 필터링 

```python
# sector가 운송인 것만 가져오기
my_concat_df.query('Sector == "운송"')
```

```python
# rename을 사용해서 컬럼명 변경
my_concat_df = my_concat_df.rename(columns={
    '2020-10-13' : '주가_10_13',
    '2020-10-14' : '주가_10_14'
})
my_concat_df
```

```python
# 주가가 2만원 넘는 것만 가져오기
my_concat_df.query('주가_10_13 > 20000')
```

```python
# 두 조건 같이
my_concat_df.query('주가_10_13 > 20000 and Sector == "제약_및_바이오"')
```

```python
# 데이터 프레임 정보 출력
my_concat_df.info()
```

```python
# 데이터 프레임 수치 정보 출력
my_concat_df.describe()
```

```python
# 값 일괄 변경
# inplace True 필수 
my_concat_df['Sector'].replace({'자동차_및_부품' : '자동차 및 부품'}, inplace=True)
my_concat_df
```

```python
# 데이터 프레임 일부를 필터링해서 새로운 데이터프레임 생성
new_df = my_concat_df[my_concat_df['Sector'] == '운송']
new_df
```
